P1. Que sucede si usamos matplotlib para mostrar una imágen leída con openCV?  


Cuando usamos OpenCV para leer una imagen, el formato de color es BGR (azul, verde, rojo), mientras que matplotlib tiene por defecto el formato RGB (rojo, verde, azul). Al mostrar la imagen sin ajustar el formato, los colores se verán alterados. Podemos corregirlo transformando de BGR a RGB usando cv2.cvtColor(img, cv2.COLOR_BGR2RGB) antes de mostrarla con matplotlib

P5. Cree un bucle para mostrar la imágen original y el efecto de las 5 diferentes interpolaciones al escalar a la mitad su ancho y largo. Se nota la diferencia?

````python
# Configuración de interpolaciones de OpenCV
interpolaciones = [
    cv2.INTER_NEAREST,
    cv2.INTER_LINEAR,
    cv2.INTER_AREA,
    cv2.INTER_CUBIC,
    cv2.INTER_LANCZOS4
]
nombres = ['Nearest', 'Linear', 'Area', 'Cubic', 'Lanczos4']

# Crear subplots
plt.figure(figsize=(12, 8))
plt.subplot(2, 3, 1)
plt.imshow(img)
plt.title("Original")
plt.axis('off')

# Escalar la imagen con cada interpolación y mostrarla
for i, (interp, nombre) in enumerate(zip(interpolaciones, nombres), 2):
    img_scaled = cv2.resize(img, (img.shape[1] // 2, img.shape[0] // 2), interpolation=interp)
    plt.subplot(2, 3, i)
    plt.imshow(img_scaled)
    plt.title(nombre)
    plt.axis('off')

plt.show()
````

### Tranformacion Logarítmica
* s = c.log(1+r)

````python
img = cv2.imread(imgSRC, 0)

fft_2 = np.fft.fft2(img)
fft_2_abs = np.abs(fft_2)
fft_2_abs = np.fft.fftshift(fft_2_abs)
fft_2_abs_log = np.log(fft_2_abs+1)

plt.figure(figsize=(10,15))
plt.imshow(fft_2_abs_log, cmap='gray')
plt.show()
````

### Histograma de una imagen
* np.histogram(img, bins, [rango de datos])  (requiere plotear)
* plt.hist(img.ravel(), bins, [rango de datos])
* cv2.calcHist([img], [canal], mascara,[bins], [rango de datos])

#####  Normalizar CDF $ \frac{cdf*hist.max}{cdf.max} $

````python
#Numpy
img = cv2.imread("./images/rxvaca.jpeg")[..., ::-1]
img_gray = img[..., 0]

hist, bins = np.histogram(img_gray, 256, [0, 256])
plt.bar(np.arange(256), hist)
plt.show()
````

````python
#matplotlib
hist = plt.hist(img_gray.ravel(), 256, [0,256], density=True)
cdf = np.cumsum(hist[0])
cdf_norm = cdf*(hist[0].max())/(cdf.max())
plt.plot(cdf_norm)
````

````python
# openCV
color = ("r*-", "gs-", "b^-")
for i, col in enumerate(color):
    hist = cv2.calcHist([img], [i], None,[256], [0,256])
    plt.plot(hist, col)
````

### Equalización de Histograma
* cv2.equalizeHist(img) 

````python
#Numpy
img = cv2.imread("./images/rxvaca.jpeg")[..., ::-1]
img_gray = img[..., 0]

img_equ = cv2.equalizeHist(img_gray)
plt.imshow(img_equ, cmap="gray", vmin=0, vmax=255)
````

P8. Que sucede durante la equalización de histograma?

Redistribuye los valores de intensidad de una imagen para mejorar su contraste, aumenta el rango de tonos en imágenes con poco contraste, haciendo que los detalles en áreas oscuras o brillantes se vuelvan más visibles

### Equalización de Histograma adaptativa
*clahe = createCLAHE() 
*clahe.apply(img) 

* [CLAHE Histogram Eqalization](https://www.geeksforgeeks.org/clahe-histogram-eqalization-opencv/)

````python
#Numpy
img = cv2.imread("./images/rxvaca.jpeg")[..., ::-1]
img_gray = img[..., 0]

clahe = cv2.createCLAHE(clipLimit=5)
final_img = clahe.apply(img_gray) + 60

plt.imshow(final_img, cmap="gray", vmin=0, vmax=255)
````
